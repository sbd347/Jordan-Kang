```python
# -*- coding: utf-8 -*-
"""
Created on Tue Nov 26 13:47:29 2019

@author: a-alhard@microsoft.com

This is a simple script, written for internal usage at Lyngby that extracts
data from QCoDeS database to a .csv file, which can later be used to
investigate, compare and manipulate. Note that the code optimized for an easy
use of extracted data within MS Excel. 
"""

from qcodes.dataset.data_set import load_by_id
from qcodes import initialise_or_create_database_at
import qcodes as qc
import os
import pandas as pd

"""
Below I obtain the location of the database from QCoDeS config.
Please change this location if you would like to use a different
database.
"""

database_location = qc.config.core.db_location
database = initialise_or_create_database_at(database_location)

def extract_data(idn: int, path_to_save: str, file_name: str): 
    data = load_by_id(idn)
    dfdict = data.get_data_as_pandas_dataframe()
    dfs_to_save = list()
    for _, df in dfdict.items():
        dfs_to_save.append(df)
    dst = os.path.join(path_to_save, f'{file_name}.csv')
    df_to_save = pd.concat(dfs_to_save, axis=1)
    df_to_save.to_csv(path_or_buf=dst, header=True, sep=',')

"""       
of the file. Note that if you have parameter data with different lengths, 
single file option cannot be use. 
"""
    
path = "C:/Users/kjh109/Desktop/My Research in NBI/Dropbox/DATA Files/PPMS Data/CSV Files/" # change this according to your desired location
name = "MQML258BB" # Give a name to the file

"""
Here, we give the absolute path where the data to be extracted and the name
"""
#id_of_data = 92 # the run id of the data to be extracted
#extract_data(id_of_data, path, name) # Run the extraction

"""
We can extract a range of runs in a db via a simple for loop.
"""
for i in range(1, 110): # Runs between 1-109
    name_of_each_run = f'MQML258BB_id {i}'
    extract_data(i, path, name_of_each_run)
```
